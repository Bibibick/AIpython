```python
# 문제 4: 결측치 확인하기
# 데이터프레임에서 결측치가 있는지 확인하고, 각 열의 결측치 개수를 출력하세요
print("문제4 : 각 열의 결측치 확인")
# display(df.isna())

print(df.isna().sum())
```

```python
# 문제 5: 이상치 확인하기
# 데이터프레임에서 이상치가 있는지 확인하세요.
# (점수 > 100 또는 공부시간 > 10 인 데이터) (조건식으로 확인)
print("문제5 : 이상치 확인(조건식)")
print("이상치 개수 :",len(df[(df['점수'] > 100) | (df['공부시간'] > 10)]))
print("이상치 데이터 : ")
display(df[(df['점수'] > 100) | (df['공부시간'] > 10)])
```

```python
# 문제 6: 중복 데이터 확인하기
# 데이터프레임에서 중복된 행이 있는지 확인하세요
print("문제6 : 완전 중복 행 확인")
print("중복 제거 전 데이터프레임 크기 :", df.shape)
duplicated_rows = df[df.duplicated(keep=False)]
print("완전 중복 행 개수 :",len(duplicated_rows))
display(duplicated_rows)
```

```python
# 문제 7: 중복 데이터 삭제하기
# 데이터프레임에서 중복된 데이터를 삭제하고, 중복이 제거된 데이터프레임을 출력하세요
print("문제7 : 완전 중복 제거")
df_no_dup = df.drop_duplicates()
print("중복 제거 후 데이터프레임 크기 :", df_no_dup.shape)
# duplicated_rows = df[df.duplicated]
print("제거된 행 수 :", len(df) - len(df_no_dup))
display(df_no_dup)
```

```python
# 문제 9: 시각화
# 공부시간과 점수에 대한 산점도 그래프를 작성하시오
plt.figure(figsize=(12, 6))
plt.scatter(df['공부시간'], df['점수'], alpha = 0.7)
plt.title('원본 데이터 : 공부시간vs점수', fontsize = 14)
plt.xlabel('공부시간', fontsize = 12)
plt.ylabel('점수', fontsize = 12)
plt.grid(True, alpha = 0.3)
```

```python
Q1_score = df['점수'].quantile(0.25)
Q3_score = df['점수'].quantile(0.75)
IQR_score = Q3_score - Q1_score
```

```python
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score
```

```python
# 문제 10: 이상치 데이터 탐색 및 제거
# 데이터프레임에서 점수와 공부시간의 이상치를 처리하세요.
# 사분위수를 이용한 판별과 제거
Q1_score = df['점수'].quantile(0.25)
Q3_score = df['점수'].quantile(0.75)
IQR_score = Q3_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

score_outliers = df[(df['점수'] < lower_bound_score) | (df['점수'] > upper_bound_score)]

Q1_study = df['공부시간'].quantile(0.25)
Q3_study = df['공부시간'].quantile(0.75)
IQR_study = Q3_study - Q1_study

lower_bound_study = Q1_study - 1.5 * IQR_study
upper_bound_study = Q3_study + 1.5 * IQR_study

study_outliers = df[(df['공부시간'] < lower_bound_study) | (df['공부시간'] > upper_bound_study)]
print('점수 이상치 :')
display(score_outliers)
print('공부시간 이상치 : ')
display(study_outliers)


df_no_outliers = df[(df['점수'] >= lower_bound_score) & (df['점수'] <= upper_bound_score)&
                    (df['공부시간'] >= lower_bound_study) & (df['공부시간'] <= upper_bound_study)]

df_no_outliers = df_no_outliers.reset_index()
print("이상치가 제거된 데이터 프레임 : ")
display(df_no_outliers)
```

```python
# 문제 11: 데이터 그룹핑
# 데이터프레임을 '분반' 열을 기준으로 그룹핑하고, 각 그룹의 '점수' 열의 평균을 계산하세요

print("문제 11 : 분반별 점수 평균")
class_mean = df_no_outliers.groupby('분반')['점수'].mean()
print(class_mean.round(2))
```

```python
# 문제 12: 데이터 시각화
# 각 '분반'의 평균 점수를 막대 그래프로 시각화하세요


plt.figure(figsize = (10, 6))
plt.bar(class_mean.index, class_mean.values, color = 'skyblue', width = 0.6)
plt.title('분반별 평균 점수', fontsize = 16)
plt.xlabel('분반', fontsize = 14)
plt.ylabel('평균 점수', fontsize = 14)
plt.grid(axis = 'y', alpha = 0.3)
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(df_no_outliers['공부시간'], df_no_outliers['점수'], alpha = 0.7)
plt.title('공부시간과 점수의 관계', fontsize = 16)
plt.xlabel('공부시간', fontsize = 14)
plt.ylabel('점수', fontsize = 14)
plt.grid(True, alpha = 0.3)
```
