```python
# 문제 2: 1부터 10까지의 숫자 중 짝수만 포함하는 배열을 생성하세요.
b = np.arange(2, 11, 2)
print(b)
```

```python
# 문제 3: [2, 4, 6] 배열과 [10, 20, 30] 배열을 각각 생성하고, 곱셈 연산을 수행한 결과를 출력하세요.
a = np.array([2, 4, 6])
b = np.array([10, 20, 30])
# c = np.array(a)
# d = np.array(b)
c = a*b
print(c)
```

```python
# 문제 4: 위 배열에서 20보다 큰 요소만 필터링해서 출력하세요.
print(c[c>20])
```
