```python
import matplotlib.pyplot as plt

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
plt.figure(figsize=(12,6))
plt.plot(month, rainfall, color='c', marker = 'o', linestyle='-.')
plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall(mm)')
plt.ylim(0, max(rainfall) * 1.1)

for i, value in enumerate(rainfall):
  plt.text(i, value + 10, str(value), ha = 'center', va = 'bottom') # 세 번째는 뿌려주는 데이터값을 말함 네 번째 다섯 번째는 위치를 말함 h는 수평(horizental) v는 수직(vertical) 고로 수평 중앙 정렬
# for i in range(len(rainfall)):
#   plt.annotate(rainfall[i], xy=(month[i],rainfall[i]), xytext=(5,0), textcoords='offset points')

plt.tight_layout()
```
